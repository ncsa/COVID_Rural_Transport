```python
import geopandas as gp
tracts = gp.read_file('tl_2019_17_tract.shp')
```

```python
import pandas as pd
pop = pd.read_csv('IL_Population.txt', dtype = str)
pop.columns = ["STATEFP", "COUNTYFP", "TRACTCE", "POPULATION","LATITUDE", "LONGITUDE"]

# pop_combined = pop.groupby('TRACTCE')
# pop = pd.DataFrame(data = pop_combined.POPULATION.sum())
pop['GEOID'] = pop[['STATEFP', 'COUNTYFP', 'TRACTCE']].agg(''.join, axis=1)
pop = pop.loc[:,('GEOID','POPULATION')]
pop.to_csv('Population_with_GEOIDs.csv', index = False)
```

The State of Illinois has 3123 census tracts. 


```python
geo_tracts = tracts.loc[:,('GEOID','geometry')]
geo_tracts_with_pop = geo_tracts.merge(pop, on = 'GEOID')
geo_tracts_with_pop.to_file('tracts_with_pop.shp')
# geo_tracts_with_pop[geo_tracts_with_pop.duplicated(subset = 'TRACTCE', keep = False).values== True]
geo_tracts_with_pop.plot(column = 'POPULATION', cmap='OrRd', scheme='quantiles')
#matplotlib.pyplot.savefig('census_tracts_with_population_heatmap.png', dpi = 300)
```

```python
from csv import reader
count = {}
with open('markets_with_zipcodes.csv', 'r') as read_obj:
    csv_reader = reader(read_obj)
    for row in csv_reader:
        if row == ['','']:
            continue
        
        locations = row[-1]
        for location in locations.strip('[]').split(','):
            zip_code = location.strip().strip("'")
            
            if zip_code not in count:
                count[zip_code] = 1
            else:
                count[zip_code] = count[zip_code] + 1
```

```python
marketCount = []
for key in count:
    row = []
    row.append(key)
    row.append(count[key])
    marketCount.append(row)
```

```python
IL_zipcodes = IL_zipcodes.loc[:,('GEOID','ZIP_CODES')]
# pop_combined = pop.groupby('TRACTCE')
# pop = pd.DataFrame(data = pop_combined.POPULATION.sum())
combined_IL_zipcodes = IL_zipcodes.groupby('GEOID')
il_zip_clean = pd.DataFrame(combined_IL_zipcodes.ZIP_CODES)
il_zip_clean.loc[:,1] = il_zip_clean.loc[:,1].apply(lambda x : (x.values))
il_zip_clean.columns = ['GEOID','ZIP_CODES']
il_zip_clean
```

```python
storeCount = []
for i in range(len(il_zip_clean)):
    stores = 0
    for j in range(len(il_zip_clean.ZIP_CODES.iloc[i])):
        zip_code = il_zip_clean.ZIP_CODES.iloc[i][j]
        if str(zip_code) in count:
            stores = stores + count[str(zip_code)]
    storeCount.append(stores)
```

```python
storedf = pd.DataFrame(storeCount)
zip_and_stores = il_zip_clean.join(storedf)
zip_and_stores.columns = ['GEOID','ZIP_CODES','STORES']
zip_and_stores
```

```python
updated = geo_tracts_with_pop.join(zip_and_stores.set_index('GEOID'), on = 'GEOID')
```

```python
updated = updated.fillna(0)
```
